```python
column_list = []
with open("communities.names", "r") as file:
    lines = file.readlines()
    specific_lines = [lines[i] for i in range(75, 203)]
    for string in specific_lines:
        column_list.append(string.split()[1])
    print(column_list)
```

```python
with open('communities.data') as f:
    data = f.readlines()

df = pd.DataFrame(data, columns=['data'])
```

```python
communities_crimes_df = pd.DataFrame()
communities_crimes_df[column_list] = df['data'].str.split(',', expand=True)
```
